# Ans.1--- header titles of Wikipedia main page

```python
# importing all the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# sending request to the web page to get the required information
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page.content

soup=BeautifulSoup(page.content)
# extracting all the header tags
# fetching all the header tiles from the website
titles=soup.find_all(['h1','h2','h3','h4','h5','h6']) 
titles
```

```python
# creating empty list for store the header titles
header_titles=[]
for i in titles:
    header_titles.append(i.text.replace('\n',''))
header_titles
```

```python
# putting all the header titles in a dta frame
headers=pd.DataFrame({}) 
headers['Header Titles of Wikipedia main page']=header_titles
```

```python
headers # final output
```

# Ans.2---- A python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release).

```python
# importing all the required libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# sending request to the web page to get the required information
page=requests.get("https://www.imdb.com/list/ls091520106/")
page.content

soup=BeautifulSoup(page.content)
# extracting all the header tags
# fetching all the header tiles from the website
titles=soup.find_all('h3',class_="lister-item-header")
titles
```

```python
ratings=soup.find_all('div',class_="ipl-rating-star small")
ratings
```

```python
movie_titles=[]
for i in titles:
    movie_titles.append(i.text.replace('\n','').replace('(', '').replace(')', ''))
movie_titles
```

```python
ratings_star=[]
for i in ratings:
    ratings_star.append(i.text.strip())
ratings_star
```

```python
release=soup.find_all('span',class_="lister-item-year text-muted unbold")
release
```

```python
release_date=[]
for i in release:
    release_date.append(i.text.replace('(', '').replace(')', ''))
release_date
```

```python
print(len(movie_titles),len(ratings_star),len(release_date))
```

```python
top=pd.DataFrame({})
top['title of the movie']=movie_titles
top['ratings of the movie']=ratings_star
top['release date']=release_date
```

```python
top
```
